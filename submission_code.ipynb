```python
def reduced1(pattern_list, text):
    match1 = re.findall("above", text)
    match2 = re.findall("over", text)
    match3 = re.findall("\+", text)
    match4 = re.findall("plus", text)
    match5 = ""
    matchbps = re.findall(r"(\d+) bps", text, re.IGNORECASE) + re.findall(r"(\d+) bpps", text, re.IGNORECASE) + re.findall(r"(\d+) bp", text, re.IGNORECASE) + re.findall(r"(\d+) basis", text, re.IGNORECASE)
    if matchbps != []:
        matchbps[0] = float(matchbps[0])
        matchbps[0] = (matchbps[0])/100
       
    for pattern in pattern_list.keys():
        if re.findall(pattern, text,re.IGNORECASE) != []:
            if matchbps != []:
                if match1 != [] or match2 != [] or match3 != [] or match4 != []:
                    match5 = pattern + "+" + str(matchbps[0])
                else:
                    match5 = pattern + "-" + str(matchbps[0])
    return match5
```

```python
def reduced2(pattern_list, text):
    match1 = re.findall(r"([-+]?\d*\.\d+|\d+)%", text) + re.findall(r"([-+]?\d*\.\d+|\d+) %", text) + re.findall(r"([-+]?\d*\.\d+|\d+) percent", text, re.IGNORECASE) + re.findall(r"([-+]?\d*\.\d+|\d+) pa", text, re.IGNORECASE) + re.findall(r"([-+]?\d*\.\d+|\d+) per annum", text, re.IGNORECASE) + re.findall(r"([-+]?\d*\.\d+|\d+) p.a", text, re.IGNORECASE) + re.findall(r"([-+]?\d*\.\d+|\d+) p.a.", text, re.IGNORECASE)
    maxv = 0
    max2v = 0
    for x in match1:
        try:
            y = float(x)
        except:
            y=float(x[1:])
        if y >= 25:
            match1.remove(x)
        if y < 7:
            match1.remove(x)
    for x in match1:
        x = float(x)
        if x > maxv:
            max2v = maxv
            maxv = x
    match2 = re.findall("above", text,re.IGNORECASE)
    match3 = re.findall("over", text,re.IGNORECASE)
    match4 = re.findall("\+", text)
    match5 = re.findall("plus", text,re.IGNORECASE)
    if match2 != [] or match3 != [] or match4 != [] or match5 != []:
        return maxv
    match6 = re.findall("below", text,re.IGNORECASE)
    match7 = re.findall("minus", text,re.IGNORECASE)
    match8 = re.findall("\-", text)
    if match6 != [] or match7 != [] or match8 != []:
        return max2v
    return maxv
```

```python
def reduced3(pattern_list, text):
    match1 = re.findall(r"([-+]?\d*\.\d+|\d+)%", text) + re.findall(r"([-+]?\d*\.\d+|\d+) %", text) + re.findall(r"([-+]?\d*\.\d+|\d+) percent", text, re.IGNORECASE) + re.findall(r"([-+]?\d*\.\d+|\d+) pa", text, re.IGNORECASE) + re.findall(r"([-+]?\d*\.\d+|\d+) per annum", text, re.IGNORECASE) + re.findall(r"([-+]?\d*\.\d+|\d+) p.a", text, re.IGNORECASE) + re.findall(r"([-+]?\d*\.\d+|\d+) p.a.", text, re.IGNORECASE)
    match7 = re.findall(r"([-+]?\d*\.\d+|\d+)%", text) + re.findall(r"([-+]?\d*\.\d+|\d+) %", text) + re.findall(r"([-+]?\d*\.\d+|\d+) percent", text, re.IGNORECASE) + re.findall(r"([-+]?\d*\.\d+|\d+) pa", text, re.IGNORECASE) + re.findall(r"([-+]?\d*\.\d+|\d+) per annum", text, re.IGNORECASE) + re.findall(r"([-+]?\d*\.\d+|\d+) p.a", text, re.IGNORECASE) + re.findall(r"([-+]?\d*\.\d+|\d+) p.a.", text, re.IGNORECASE)
    match2 = re.findall("above", text,re.IGNORECASE)
    match3 = re.findall("over", text,re.IGNORECASE)
    match4 = re.findall("\+", text,)
    match5 = re.findall("plus", text,re.IGNORECASE)
    match6 = ""
    minv = 7
    for x in match1:
        if x == "3":
            match1.remove(x)
        if x == "6":
            match1.remove(x)
        try:
            y = float(x)
        except:
            y = float(x[1:])
        if y >= 7:
            match1.remove(x)
        if y <= -7:
            match1.remove(x)
    for x in match1:
        try:
            y = float(x)
        except:
            y = float(x[1:])
        if y < minv:
            minv = y
    if minv >= 7:
            minv = 0
    matchint = minv
    for pattern in pattern_list.keys():
        if re.findall(pattern, text,re.IGNORECASE) != []:
            if matchint < 0:
                match6 = pattern + str(matchint)
            elif match2 != [] or match3 != [] or match4 != [] or match5 != []:
                match6 = pattern + "+" + str(matchint)
            else:
                match6 = pattern + "-" + str(matchint)
    return match6
```
